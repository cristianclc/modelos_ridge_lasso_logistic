# Información del Dataset

## Contexto

Craigslist es la mayor colección de vehículos usados en venta del mundo; sin embargo, es muy difícil reunirlos todos en un mismo lugar. Entonces, se creo este conjunto de datos que incluye todas las entradas de vehículos usados en Estados Unidos en Craigslist.

## Contenido

Estos datos se extraen cada pocos meses y contienen casi toda la información relevante que Craigslist proporciona sobre las ventas de automóviles, incluyendo columnas como precio, estado, fabricante, latitud/longitud y otras 18 categorías.

## Dataset

Link del Dataset: https://www.kaggle.com/datasets/austinreese/craigslist-carstrucks-data

## Lista de Variables

- `id`

- `url`

- `region`: región de craigslist.

- `region_url`: URL de la región.

- `price`: Precio del vehículo.

- `year`: Año del vehículo.

- `manufacturer`: Marca de vehículo.

- `model`: Modelo del vehículo.

- `condition`: Condición del vehículo.

- `cylinders`: Número de cilindros del vehículo.

- `fuel`: Tipo de combustible del vehículo.

- `odometer`: Millas recorridas del vehículo.

- `title_status`: Estado del título del vehículo.

- `transmission`: Tipo de transmisión del vehículo.

- `VIN`: Número de identificación del vehículo.

- `drive`: Tipo de tracción del vehículo.

- `size`: Tamaño del vehículo.

- `type`: Tipo genérico del vehículo.

- `paint_color`: Color del vehículo.

- `image_url`

- `description`: Descripción del vehículo.

- `county`: Columna inútil dejada por error.

- `state`: Estado de cotización.

- `lat`: Latitud de cotización

- `long`: Longitud del listado.

- `posting_date`
