# Información del Dataset

## ¿Con qué propósito se creó el conjunto de datos?
El conjunto de datos se creó en un proyecto que busca contribuir a la reducción del abandono y el fracaso académico en la educación superior, mediante el uso de técnicas de aprendizaje automático para identificar a los estudiantes en riesgo en una etapa temprana de su trayectoria académica, de modo que se puedan implementar estrategias de apoyo.

El conjunto de datos incluye información conocida al momento de la matriculación del estudiante: trayectoria académica, datos demográficos y factores socioeconómicos.

El problema se formula como una tarea de clasificación de tres categorías (abandono, matriculado y graduado) al final de la duración normal del curso.

## ¿Quién financió la creación del conjunto de datos?
Este conjunto de datos cuenta con el apoyo del programa SATDAP - Capacitação da Administração Pública bajo la subvención POCI-05-5762-FSE-000191, Portugal.

## ¿Qué representan las instancias de este conjunto de datos?
Cada instancia es un estudiante.
